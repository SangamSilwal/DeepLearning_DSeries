# RNN (Recurrent Neural Networks)
A Recurrent Neural Network (RNN) is a type of artificial neural network designed to process sequential data or data with temporal dependencies, such as time series, text, or speech.

```python
import numpy as np
docs = [
    'sangam is my name',
    'sugam is his name',
    'jai nepal',
    'jai nepal aama',
    'kp sharma oli',
    'paras khadka',
    'paras pariyar'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequence = tokenizer.texts_to_sequences(docs)
sequence
```

```python
from keras.utils import pad_sequences
sequence = pad_sequences(sequence,padding='post')
sequence
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train.shape
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train[6]
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
