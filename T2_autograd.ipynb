```python
import torch
```

```python
x = torch.tensor([1],requires_grad=True,dtype=torch.float32)
y = x**2
# To calculate the gradient we need to write require_grad=True
```

```python
x
```

```python
y
```

```python
z = torch.sin(y)
```

```python
z.backward()
```

```python
z
```

```python
x.grad
```

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0,requires_grad=True)
b = torch.tensor(0.0,requires_grad=True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
```

```python
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
```

```python
loss = F.binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
b.grad
```

```python
x = torch.tensor([1.0,2.0,3.0],requires_grad=True)
```

```python
y = (x**2).mean()
y
```

```python
y.backward()
x.grad
```

```python
# Clearing Grad
x = torch.tensor(2.0,requires_grad=True)
x
```

```python
y = x**2
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

```python
x.grad.zero_()
```

```python
# Disable Gradient tracking
x.requires_grad_(False)
```

```python
x
```

```python
# Detach
x = torch.tensor(2.0,requires_grad=True)
x
```

```python
z = x.detach()
```

```python
z
```

```python
# torch.no_grad -> function
x = torch.tensor(1.0,requires_grad=True)
x
```

```python
with torch.no_grad():
  y = x**2
```

```python
y
```
