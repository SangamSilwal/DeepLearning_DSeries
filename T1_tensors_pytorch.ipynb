```python
import torch
print(torch.__version__)
```

# Creating Tensors

```python
# using empty
a = torch.empty(2,3)
print(type(a))
a
```

```python
#using zero
torch.zeros(2,3)
```

```python
# using ones
torch.ones(2,3)
```

```python
# using rand
torch.rand(2,3)
```

```python
# use of seed
torch.manual_seed(100)
print(torch.rand(2,3))
print(torch.rand(2,3))
```

```python
torch.manual_seed(100)
print(torch.rand(2,3))
print(torch.rand(2,3))
```

```python
# using tensor
torch.tensor([[[1,2,3],[2,3,4]]]).shape
```

```python
#other ways to create tensor

# arange
print("using arange-> ",torch.arange(0,10,2))

# using linespace
print("using linespace-> ",torch.linspace(0,10,10))

# using eye
print("using eye-> ",torch.eye(5,5))  #identity matrix

# using full
print("using full-> ",torch.full((3,3),6))
```
