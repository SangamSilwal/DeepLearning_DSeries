```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Churn_Modelling.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True)
```

```python
df.head(2)
```

```python
df = pd.get_dummies(df,columns=["Geography","Gender"],drop_first=True)
```

```python
df.head(2)
```

```python
df['Geography_Germany']=np.where(df['Geography_Germany']==1,1,0)
df['Geography_Spain']=np.where(df['Geography_Spain']==1,1,0)
```

```python
df['Gender_Male'] = np.where(df['Gender_Male']==1,1,0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["Exited"]),df["Exited"],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(11,activation="relu",input_dim=11))
model.add(Dense(11,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train,epochs=100,verbose=1,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
